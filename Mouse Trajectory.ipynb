```python
with open (filename, 'r') as f:
    i = 0
    for line in f:
        tokens=line.split(',')
        t, x, y = [float(x) for x in tokens]
        m_traj[i][0]=t
        m_traj[i][1]=x
        m_traj[i][2]=y
        i+=1
```

```python
m_traj[0]
```

```python
mtraj = m_traj.copy()
```

```python
t0=m_traj[0][0]
for i in xrange(mtraj.shape[0]):
    mtraj[i][0] = mtraj[i][0]-t0
```

```python
from scipy import interpolate
```

```python
tnew = np.arange(0, 10, .001)
xtck = interpolate.splrep(mtraj.T[0], mtraj.T[1], s=0)
ytck = interpolate.splrep(mtraj.T[0], mtraj.T[2], s=0)
xnew = interpolate.splev(tnew, xtck, der=0)
ynew = interpolate.splev(tnew, ytck, der=0)
```

```python
plt.plot(tnew, xnew)
plt.plot(tnew, ynew)
```
